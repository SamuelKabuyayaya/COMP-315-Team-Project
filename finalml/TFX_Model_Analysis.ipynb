```python
eval_result_3= tfma.load_eval_result(
    '/home/manjula/COMP315/airflow_pipeline_outputs/adult_census_tfx_output/Evaluator/evaluation/94'
)
```

```python
m1_slices = eval_result_1.get_metrics_for_all_slices()
m2_slices = eval_result_2.get_metrics_for_all_slices()
m3_slices = eval_result_3.get_metrics_for_all_slices()

for slice_key, m1 in m1_slices.items():
    if slice_key not in m2_slices or slice_key not in m3_slices:
        continue  # skip slices that don't exist in all three runs

    m2 = m2_slices[slice_key]
    m3 = m3_slices[slice_key]

    acc1 = m1.get('binary_accuracy', {})
    acc2 = m2.get('binary_accuracy', {})
    acc3 = m3.get('binary_accuracy', {}) 
    auc1 = m1.get('auc', {})
    auc2 = m2.get('auc', {})
    auc3 = m3.get('auc', {})              

    print(f"\nSlice: {slice_key}")
    print(f"  Run 69 accuracy: {acc1}")
    print(f"  Run 78 accuracy: {acc2}")
    print(f"  Run 94 accuracy: {acc3}")   
    print(f"  Run 69 AUC:      {auc1}")
    print(f"  Run 78 AUC:      {auc2}")
    print(f"  Run 94 AUC:      {auc3}")   
```
