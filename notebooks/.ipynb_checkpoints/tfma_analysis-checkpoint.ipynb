```python
import os
import tensorflow_model_analysis as tfma
from pathlib import Path

# paths
PROJECT_ROOT = Path("/home/samso/comp315_project")
EVAL_BASE_DIR = PROJECT_ROOT / "tfx_pipeline_output_v2" / "Evaluator" / "evaluation"

try:
    all_runs = [int(d.name) for d in EVAL_BASE_DIR.iterdir() if d.is_dir() and d.name.isdigit()]
    if not all_runs:
        print("No folders with number sound, check here:", list(EVAL_BASE_DIR.iterdir()))
        raise FileNotFoundError
        
    latest_run = max(all_runs)
    EVAL_DIR = EVAL_BASE_DIR / str(latest_run)
    
    print(f"found latest run: {latest_run}")
    
    result = tfma.load_eval_result(output_path=str(EVAL_DIR))
    print(" Data loaded into 'result'")
    
except Exception as e:
    print(f"error loading eval result: {e}")
```
